# Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv('/home/yoon/Downloads/weather_data_20190101_20191231_num_filtered.csv')
weather
```

```python
rain = weather.groupby("TM")["RN_DAY"].sum().reset_index()
rain = rain["RN_DAY"]
rain
```

### 이 밑에가 일별로 하나씩 남겨놓은 거. RN day는 sum, 나머지는 average로 처리

```python
weather_average = weather.groupby('TM').mean().reset_index()
weather_average['TM'] = pd.to_datetime(weather_average['TM'], format='%Y%m%d')
weather_average["RN_DAY"] = rain
weather_average
```

```python
weather_average.to_csv("weather_average2019.csv", index=False)
```
